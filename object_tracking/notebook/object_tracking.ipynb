```python
def is_extracted(save_path):
    with open(save_path, 'rb') as handle:
        data = pickle.load(handle)
    track_id = save_path.split('/')[-1].split('.')[0]
    keys = list(data[track_id].keys())
    if (data is None) or (len(keys) == 0):
        return False
    return True
```

```python
import pickle
keys = []
veh_dir = '/content/nanonets_object_tracking/feature/vehicle'
color_dir = '/content/nanonets_object_tracking/feature/color'

for key in tqdm(track_keys_color):
    path = os.path.join(color_dir, key)
    if is_extracted(path):
        keys.append(key)

for key in tqdm(track_keys_vehicle):
    path = os.path.join(veh_dir, key)
    if is_extracted(path):
        keys.append(key)
```
